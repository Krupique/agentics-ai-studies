```python
import yfinance as yf
import matplotlib.pyplot as plt

ticker = yf.Ticker("BTLG11.SA")

# Histórico de preços
preco = ticker.history(period="6mo")

# Dividendos
dividendos = ticker.dividends

print("Últimos preços:")
print(preco.tail())

print("\nÚltimos dividendos:")
print(dividendos.tail())

# Gráfico de preço
preco["Close"].plot(title="BTLG11 - Preço")
plt.show()
```
